```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import math
```

```python
# read the clean data csv file
data = pd.read_csv('/content/clean_diabetes.csv')
```

```python
# display first 10 columns
data.head(10)
```

```python
# drop the Unmamed column
data = data.drop(data.columns[0], axis=1)
data.head(10)
```

```python
# get a random sample of 25 observations
data_sample = data.sample(25, random_state=0)
data_sample
```

# Part A

```python
# Part A
# calculate the mean Glucose of a random sample of 25 and the population
sample_mean_Glucose = data_sample['Glucose'].mean()
population_mean_Glucose = data['Glucose'].mean()

# creating the bar plot
means = [sample_mean_Glucose, population_mean_Glucose]
values = ['sample_25', 'population']
fig = plt.figure()
plt.bar(values, means, color=['red','blue'])
plt.title("Mean Glucose of a sample VS Mean Glucose of the population")
plt.show()
```

The mean Glucose of a random sample of 25 observations is very close to the mean Glucose of the population





```python
# calculate the max Glucose of a random sample of 25 and population
sample_max_Glucose = data_sample['Glucose'].max()
population_max_Glucose = data['Glucose'].max()

# creating the bar plot
maxs = [sample_max_Glucose, population_max_Glucose]
values = ['sample_25', 'population']
fig = plt.figure()
plt.bar(values, maxs, color=['green','orange'])
plt.title("Max Glucose of a sample VS Max Glucose of the population")
plt.show()
```

The max Glucose of a random sample of 25 observations is as same as the highest
Glucose of the population

# Part B

```python
# Part B
# calculate the 98 percentile of BMI of a random sample of 25 and the population
sample_98_percentile_BMI = data_sample['BMI'].quantile(0.98)
population_98_percentile_BMI = data['BMI'].quantile(0.98)
 
# creating the bar plot
percentile_98 = [sample_98_percentile_BMI, population_98_percentile_BMI]
values = ['sample_25', 'population']
fig = plt.figure()
plt.bar(values, percentile_98, color=['lime','gray'])
plt.title("98 percentile of BMI of a sample compared with the population")
plt.show()
```

The 98 percentile of BMI of a random sample of 25 observations is less than the 98 percentile of BMI of the population

# Part C

```python
# population distrubution of BloodPressure
fig,ax = plt.subplots()
ax.hist(data['BloodPressure'], bins=30, ec='b')
plt.title("Population distrubution of BloodPressure")
plt.show()
```

```python
# create 500 samples (of 150 observation each) from the population and plot the sampling distribution of the mean for BloodPressure
sample_means = []
sample_size = 150

for i in range(500):  # do the following 500 times
    sample = data.sample(sample_size,replace=True)  # take a sample of 150 values from the population
    sample_means.append(np.mean(sample['BloodPressure']))  # add the mean of the sample's BloodPressure to the sample_means list

# creating the bar plot
fig,ax = plt.subplots()
ax.hist(sample_means, bins=30, ec='b')
plt.title('The sampling distribution of the mean for BloodPressure')
plt.show()
```

The shape of the sampling distribution of the mean tends to be bell-shaped

```python
# caculate the average mean of 500 samples and population mean for BloodPressure
mean_sample_means_BP = np.mean(sample_means)
population_mean_BP = data['BloodPressure'].mean()

# creating the bar plot
mean_values = [mean_sample_means_BP, population_mean_BP]
values = ['500_samples', 'population']
fig = plt.figure()
plt.bar(values, mean_values, color=['red','blue'])
plt.title("The average mean of 500 samples VS population mean for BloodPressure")
plt.show()
```

The average mean of 500 samples is very close to the population mean for BloodPressure

```python
# caculate the standard derviation of the mean of 500 samples and population standard derviation for BloodPressure
std_samples_BP = np.std(sample_means)
population_std_BP = data['BloodPressure'].std()
standard_error_BP = population_std_BP/math.sqrt(sample_size)

# creating the bar plot
std_values = [standard_error_BP, population_std_BP]
values = ['500_samples', 'population']
fig = plt.figure()
plt.bar(values, std_values, color=['lime','gray'])
plt.title("The sd of the mean of 500 samples VS population sd for BloodPressure")
plt.show()
```

The standard error is very small which means that sample means are closely distributed around the population mean.



```python
# calculate the 98 percentile of BloodPressure of the mean of 500 samples and the population
samples_98_percentile_BP = np.quantile(sample_means,0.98)
population_98_percentile_BP = data['BloodPressure'].quantile(0.98)
 
# creating the bar plot
percentile_98 = [samples_98_percentile_BP, population_98_percentile_BP]
values = ['500_samples', 'population']
fig = plt.figure()
plt.bar(values, percentile_98, color=['red','blue'])
plt.title("98 percentile of BP of 500 samples compared with the population")
plt.show()
```

The 98 percentile of BloodPressure of the mean of 500 samples is less than the 98 percentile of BloodPressure of the population
